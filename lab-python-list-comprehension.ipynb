````python
inventory = {}
products = ["t-shirt", "mug", "hat", "book", "pen"]
````

````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory
````

````python
inventory = initialize_inventory(products)
````

2. Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension.


````python
def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders "))
    customer_orders = [input("Enter a product to order: ") for product in range(num_orders)]

    return customer_orders
````

````python
orders = get_customer_orders()
print(orders)
````

3. Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

````python
def total_price(orders):
    prices = [float(input("Enter the price of the product: ")) for product in orders] 
        #cuidado, float --> porque necesitamos que sea un número lo que pone el cliente (por defecto es str), y además, el paréntesis debe ir antes de "for..."
    return sum(prices)  #Sum of all of the prices (calculate the total price)
````

````python
prices = total_price(orders)
print(prices)
````

4. Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero after fulfilling the customer orders. Use comprehension to filter out the products with a quantity of zero from the inventory.

````python
def update_inventory(orders, inventory):
    inventory = {product: (quantity - 1 if product in orders and quantity >0 else quantity) for product, quantity in inventory.items()}  #-1 product if ordered
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}  #Remove the product if = 0 after ordering

    return inventory
````

````python
inventory = update_inventory(orders, inventory)
````

````python
print(inventory)
````
